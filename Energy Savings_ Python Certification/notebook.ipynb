```python
import pandas as pd

#First dataframe
df2003 = pd.read_csv('datasets/wastestats.csv')
df2003.loc[df2003['waste_type']=='Plastics','waste_type']='Plastic'
df2003.loc[df2003['waste_type'].isin(['Ferrous metal','Ferrous Metal']),'waste_type']='Ferrous Metals'
df2003.loc[df2003['waste_type'].isin(['Non-ferrous metal','Non-ferrous metals']),'waste_type']='Non-ferrous Metals'
df2003 = df2003[df2003['waste_type'].isin(['Plastic','Ferrous Metals','Non-ferrous Metals','Glass'])]
#df2003['waste_type'].unique()
display(df2003.head())
```

```python
#Second dataframe
df2018 = pd.read_csv('datasets/2018_2019_waste.csv')
df2018.columns=['waste_type','total_waste_generated_tonne','total_waste_recycled_tonne','year']
df2018['total_waste_generated_tonne'] = df2018['total_waste_generated_tonne']*1000
df2018['total_waste_recycled_tonne'] = df2018['total_waste_recycled_tonne']*1000
df2018.loc[df2018['waste_type']=='Plastics','waste_type']='Plastic'
df2018.loc[df2018['waste_type']=='Ferrous Metal','waste_type']='Ferrous Metals'
df2018.loc[df2018['waste_type']=='Non-Ferrous Metal','waste_type']='Non-ferrous Metals'
df2018 = df2018[df2018['waste_type'].isin(['Plastic','Ferrous Metals','Non-ferrous Metals','Glass'])]
#df2018['waste_type'].unique()
display(df2018)
```

```python
df = df2003.append(df2018)
df = df[['year','waste_type','total_waste_recycled_tonne']]
df = df.sort_values(by='year')
df = df[df['year']>2014]
display(df)
```

```python
dfen = pd.read_csv('datasets/energy_saved.csv')
dfen = dfen.transpose()
dfen = dfen[[2,3]]
dfen.index = [0,1,2,3,4,5]
dfen.columns = ['waste_type', 'energy_saved']
dfen = dfen.iloc[1:]
dfen.loc[dfen['waste_type']=='Ferrous Metal','waste_type']='Ferrous Metals'
dfen.loc[dfen['waste_type']=='Non-Ferrous Metal','waste_type']='Non-ferrous Metals'

display(dfen)
```

```python
finaldf = df.merge(dfen)
finaldf = finaldf.sort_values(by='year')
display(finaldf)
```

```python
finaldf['energy_saved'] = finaldf['energy_saved'].str.strip('Kwh')
finaldf['energy_saved'] = finaldf['energy_saved'].astype('float')
finaldf['total_energy_saved'] = finaldf['total_waste_recycled_tonne']*finaldf['energy_saved']
display(finaldf)
```

```python
annual_energy_savings = finaldf.groupby('year')[['total_energy_saved']].sum()
#display(annual_energy_savings)
```
